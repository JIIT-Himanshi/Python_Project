```python
df['player_of_match'].value_counts()[0:5].plot(kind='barh', color='blue', figsize=(10, 5))
plt.title('Top 5 Players of the Match')
```

```python
df.describe(include='all')
```

```python
top_players = df['player_of_match'].value_counts().head(10)
print(top_players)

# Bar plot
plt.figure(figsize=(10,5))
top_players.plot(kind='bar', color='orange')
plt.title('Top 10 Players with Most Player of the Match Awards')
plt.xlabel('Player')
plt.ylabel('No. of Awards')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['win_by_runs'], bins=30, kde=True, color='blue')
plt.title('Distribution of Wins by Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12,6))
sns.histplot(df['win_by_wickets'], bins=30, kde=True, color='green')
plt.title('Distribution of Wins by Wickets')
plt.xlabel('Wickets')
plt.ylabel('Frequency')
plt.show()
```

```python
toss_decision = df['toss_decision'].value_counts()
toss_decision.plot.pie(autopct='%1.1f%%', shadow=True, colors=['lightblue','lightgreen'])
plt.title('Toss Decision Distribution')
plt.ylabel('')
plt.show()
```

```python
top_teams = df['winner'].value_counts().head(10)
print(top_teams)

plt.figure(figsize=(10,5))
top_teams.plot(kind='bar', color='purple')
plt.title('Top Winning Teams')
plt.xlabel('Team')
plt.ylabel('Number of Matches Won')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
city_matches = df['city'].value_counts()
plt.figure(figsize=(12,6))
city_matches.plot(kind='barh', color='teal')
plt.title('Number of Matches by City')
plt.xlabel('No. of Matches')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

```python
df['Toss_Win_and_Match_Win'] = df['toss_winner'] == df['winner']
toss_effect = df['Toss_Win_and_Match_Win'].value_counts()
toss_effect.plot.pie(autopct='%1.1f%%', labels=['Yes', 'No'], colors=['#ff9999','#66b3ff'], shadow=True)
plt.title('Does Winning Toss Help Win the Match?')
plt.ylabel('')
plt.show()
```

```python
most_wins = df['winner'].value_counts().idxmax()
print(f'Based on 2024 data, the most successful team likely to win in 2025 is: {most_wins}')

```
